## 1. Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

```python
import pandas as pd
# Dataset 1: Product details
df1 = pd.DataFrame({
    'ProductID' : [101,102,103],
    'ProductName': ['Pen', 'Notebook', 'Marker'],
    'Price': [10,50,30]
})

# Dataset 2: More products (same columns as df1 → good for vertical stacking)
df2 = pd.DataFrame({
    'ProductID': [104, 105],
    'ProductName': ['Pencil', 'Eraser'],
    'Price': [5,4]
})
# Dataset 3: Additional info (different columns → good for horizontal stacking)
df3 = pd.DataFrame({
    'Stock': [100, 60, 80],
    'Rating': [4.5,4.7,4.8]
})
```

```python
print(f"DF1: {df1}")
print(f"DF2: {df2}")
print(f"DF3: {df3}")
```

```python
# Vertical staking (axis = 0)
vertical_staking = pd.concat([df1, df2], axis=0, ignore_index=True)
print(f"Vertical staking: {vertical_staking}")
```

```python
# Horizontal staking
horizontale_staking = pd.concat([df1,df3],axis=1)
# print(f"Horizontal staking: {horizontale_staking}")
horizontale_staking
```

## 2. Using `merge()` for SQL-Style Joins
- `merge()` is similar to SQL joins.
- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

```python
# You gays have to concate df1 and df2 row wise
product = pd.concat([df1, df2], ignore_index=True)
print(f"Product: {product}")
```

```python
sales = pd.DataFrame({
    'ProductID': [101,102,104,106],
    'UnitsSold' : [500,200,150,50]
})
sales
```

```python
# Inner join
inner_join = pd.merge(product, sales, how='inner')
# print(f"Inner join: {inner_join}")
inner_join
```

```python
# left join
left_join = pd.merge(product, sales, how='left')
# print(f"Left join: {left_join}")
left_join
```

```python
# right join
right_join = pd.merge(product,sales,how='right')
# print(f"Right join: {right_join}")
right_join
```

```python
# Outer join
outer_join = pd.merge(product, sales, how='outer')
# print(f"Outer join: {outer_join}")
outer_join
```
