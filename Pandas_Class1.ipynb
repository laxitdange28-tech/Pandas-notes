## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
import pandas as pd
marks = pd.Series([10,20,30,40,50],index=['Ronaldo','Messi','Virat','Dhoni','Mbappe'])

print(marks['Virat'])
````

### Creating a DataFrame

````python
df = pd.DataFrame({
    'Name':['Ronaldo','Messi','Virat','Dhoni','Mbappe'],
    'Age':[35,34,33,32,31],
    'marks':[10,20,30,40,50]
    })
print(df)

# Selecting a column
print(df['Age'])

# To get specific row we use .loc
print(df.loc[0])

# select a specific value
print(df.loc[1,'Name'])
````

### Creating DataFrame from NumPy Arrays

````python
information = pd.DataFrame([
    ['Rahul',52,80],
    ['Devraj',22,85],
    ['Rohit',60,55]

], columns=['Name','Age','Marks']
    )
print(information)

print(information.loc[1,'Marks'])

# To get specific row we use .loc
print(information.loc[1])
````

````python
# generete a 3x4 materix with random numbers(1-13) using numpy
# name the columns as [a,b,c,d]
import numpy as np
matrix = np.random.randint(1,13, size=(3,4))
print(matrix)
df = pd.DataFrame(matrix,columns=['A','B','C','D'])
print(df)
````
