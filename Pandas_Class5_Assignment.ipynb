## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
# Write your code here
df = pd.DataFrame({
    'Marks' : [60,70,80,90]
})
print("Mean = ",df.mean())
print("Median = ",df.median())
print("Max = ",df.max())
print("Min = ",df.min())
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
# Write your code here
df = pd.DataFrame({
    'Nmae' : ['Jay','Veru','Amit','Yeas','Ayush'],
    'Age' : [20,21,22,23,24],
    'Score' : [90,80,70,60,50],
    'Height' : [170,175,180,185,188]
})
print(df)
```

```python
# 1. Sort by Score (Descending)
df['Score'].sort_values(ascending=False)
```

```python
# 2. Find Mean Age
print(df['Age'].mean())
```

```python
# 3. Find Median Height
print(df['Height'].median())
```

```python
# 4. Show Score Frequency
df['Score'].value_counts()
```
