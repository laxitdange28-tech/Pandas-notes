# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
import pandas as pd
````

````python
# Write your code here
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
df = pd.DataFrame(City, columns=['City'])
df['City'].unique()
````


# Q2. Count Unique Values


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
# Write your code here
df = pd.DataFrame({
    'Name' : ['Amit','Ayush',"Ajay","Jay"],
    'Age' : [20,21,22,23],
    'Department' : ['HR','IT','HR','IT'],
    'Salary' : [10000,20000,30000,40000]
})

df.groupby('Department')['Salary'].sum()
````
